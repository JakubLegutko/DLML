## Zarządzanie obrazami


```python
### Zadanie 2
import numpy as np
from PIL import Image

## Wczytanie obrazu
img = Image.open('cat.jpg')
img = np.array(img)
## Podstawowe informacje o obrazie
print("Podstawowe informacje o obrazie")
print(img.shape)
print(img.ndim)
print(img.dtype)
print(type(img))
## Wyświetlenie obrazu
import matplotlib.pyplot as plt

## Zapis do pliku
print("Zapis do pliku")
img = Image.fromarray(img)
img.save('cat2.jpg')
## Wypisanie informacji o obrazie
print("Wypisanie informacji o obrazie")
print(img.format)
print(img.mode)
print(img.size)
print("Wyświetlenie obrazu")
plt.imshow(img)
plt.show()
## Zmiana rozmiaru obrazu
print("Zmiana rozmiaru obrazu")
img = Image.open('cat.jpg')
img = np.array(img)
img = Image.fromarray(img)
img = img.resize((100, 100))
img.save('cat3.jpg')
plt.imshow(img)
plt.show()
## Obrót obrazu
print("Obrót obrazu")
img = Image.open('cat.jpg')
img = np.array(img)
img = Image.fromarray(img)
img = img.rotate(90)
img.save('cat4.jpg')
plt.imshow(img)
plt.show()
## Konwersja do skali szarości
print("Konwersja do skali szarości")
img = Image.open('cat.jpg')
img = np.array(img)
img = Image.fromarray(img)
img = img.convert('L')
img.save('cat5.jpg')
plt.imshow(img  , cmap='gray')
plt.show()
## Wycinanie fragmentu obrazu
print("Wycinanie fragmentu obrazu")
img = Image.open('cat.jpg')
img = np.array(img)
img = Image.fromarray(img)
img = img.crop((100, 100, 300, 300))
img.save('cat6.jpg')
plt.imshow(img)
plt.show()
## Przygotowanie danych do sieci neuronowej
print("Przygotowanie danych do sieci neuronowej")
img = Image.open('cat.jpg')
img = np.array(img)
img = Image.fromarray(img)
## Skalowanie
img = img.resize((100, 100))
## Centrowanie
img = np.array(img)
img = img - np.mean(img)
## Standaryzacja
img = img / np.std(img)
## Normalizacja
img = (img - np.min(img)) / (np.max(img) - np.min(img))


```
